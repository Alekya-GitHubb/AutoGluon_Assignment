```python
import pandas as pd
import numpy as np

DATA_DIR = "/content/ieee/"

test_tr = pd.read_csv(DATA_DIR + "test_transaction.csv")
test_id = pd.read_csv(DATA_DIR + "test_identity.csv")
test = test_tr.merge(test_id, on="TransactionID", how="left")

# Add missing columns
missing_cols = [col for col in predictor.feature_metadata_in.get_features() if col not in test.columns]
for col in missing_cols:
    test[col] = np.nan

common_cols = [col for col in predictor.feature_metadata_in.get_features() if col in test.columns]
test = test[common_cols]
```

```python
import numpy as np

chunk_size = 50000  # you can reduce to 25000 if it still crashes
predictions = []

for start in range(0, len(test), chunk_size):
    end = min(start + chunk_size, len(test))
    print(f"Processing rows {start} to {end} ...")
    chunk = test.iloc[start:end]
    chunk_pred = predictor.predict_proba(chunk, as_multiclass=False)
    predictions.append(chunk_pred)

# Combine all predictions
y_pred_proba = pd.concat(predictions)
print("✅ All predictions generated successfully.")
```

```python
sub = pd.read_csv(DATA_DIR + "sample_submission.csv")
sub["isFraud"] = y_pred_proba.values
sub.to_csv("/content/my_submission.csv", index=False)
print("✅ Final submission saved at /content/my_submission.csv")

sub.head()
```
