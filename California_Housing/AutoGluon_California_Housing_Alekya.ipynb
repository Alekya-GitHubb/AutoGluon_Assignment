```python
!pip install -q autogluon
import pandas as pd
from autogluon.tabular import TabularPredictor
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing(as_frame=True)
df = data.frame

# Add the target column explicitly
df['MedHouseValue'] = data.target

print("Shape:", df.shape)
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42)
label = 'MedHouseValue'
```

```python
PERSIST_DIR = "/content/california_autogluon/"

predictor = TabularPredictor(label=label, path=PERSIST_DIR, problem_type='regression')
predictor.fit(train, presets='best_quality', time_limit=600)
```

```python
predictor.leaderboard(test, silent=True)
```

```python
y_pred = predictor.predict(test)
output = pd.DataFrame({'Actual': test[label], 'Predicted': y_pred})
output.to_csv('/content/california_predictions.csv', index=False)
output.head()
```
